### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinski Zaton) с течением времени

````python
import matplotlib.pyplot as plt
%matplotlib inline
````

2. Постройте гистограмму распределения цен на квартиры с общей площадью более 50 кв.метров

````python
market[market['full_sq'] > 50]['price_doc'].hist(bins=30, figsize=(10, 6))
plt.title('Распределение цен на квартиры >50 кв.м'); 
````

3. Визуализируйте количество проданных квартир в разных районах

````python
result = market.loc[market['product_type'] == 'OwnerOccupier', ['sub_area', 'product_type']].set_index('product_type')
plt.hist(result, bins=len(result['sub_area'].unique()))
plt.title('Количество проданных квартир по районам'); 
plt.xticks(rotation=90); 
plt.tight_layout();
````

4. Постройте box plot отображающий распределение полных площадей квартир в Басманном районе

````python
market[market['sub_area'] == 'Basmannoe']['full_sq'].plot.box(figsize=(8, 6))
plt.title('Распределение полных площадей в Басманном районе');
````

5. Визуализируйте зависимость средней стоимости квартир в зависимости от этажа

````python
market.groupby('floor')['price_doc'].mean().plot(figsize=(12, 6))
plt.title('Зависимость средней стоимости квартир от этажа')
plt.xlabel('Этаж'); plt.ylabel('Средняя цена');
````

6. Покажите как меняется средняя площадь квартир в зависимости от района

````python
district_areas = market.groupby('sub_area')['full_sq'].mean()
plt.figure(figsize=(25, 10))
plt.bar(district_areas.index, district_areas.values)
plt.title('Сравнение средней площади жилья в различных районах', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
````

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
small_apt_prices = market[(market['full_sq'] < 50) & 
                          (market['timestamp'].between('2011-01-01', '2014-12-31'))]
small_apt_prices['period'] = pd.to_datetime(small_apt_prices['timestamp']).dt.to_period('M')
price_trend = small_apt_prices.groupby('period')['price_doc'].median()

plt.figure(figsize=(12, 6))
plt.plot(price_trend.index.astype(str), price_trend.values, marker='o', linewidth=2)
plt.title('Динамика стоимости компактного жилья (2011-2014)', fontsize=14)
plt.xlabel('Временной период')
plt.ylabel('Медианная цена')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
````
